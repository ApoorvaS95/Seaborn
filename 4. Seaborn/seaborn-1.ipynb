```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #isse bss dataset import hi karenge aaj
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='total_bill', y='tip') #seaborn mein data bhi pass karna padta h
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip') #yaha label nhi paas karne padte alag se, seaborn data se hi utha leta h
```

```python
# scatter plot -> axes level function - ye poora plot one by one samjahte h
```

```python
#hue parameter --> suppose mai iss upar wale plot mai male aur female k basis pe variation bhi dekhna chahta hu
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex') #legends are automatically generated
```

```python
# style parameter - suppose hume ye dekhna h ki data mai konse observations kis time pe restaurant mai aaye
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time')
# cross --> dinner
# dot --> lunch
```

```python
#suppose ab hume size k basis pe bhi variation check karna h
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time',size='size')
```

### scatter plot - axes level function se banaya hua plot h
### relplot - figure level function se banaya hua plot h

### seaborn documentation hi recommend karti h ki aap zyada se zyada figure level function use kare

```python
# relplot -> figure level -> square shape mai aataa h, yaha kind pass karta padega
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
# hue parameter 
# suppose hume dekhna h ki konse dots male h aur konse female
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex')
```

```python
# scatter using relplot -> style and hue
# suppose hume dehna h ki konse customers lunch k time pe aaye ar konse dinner k time pe
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style = 'time')
```

```python
# size semantics
#agar hume ye dekhna h ki konsa customer kitne logo k sath aaya
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style = 'time', size='size')
```

```python
# line plot
gap = px.data.gapminder() #plotly se data utha rhe h
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
#mujhe ye dekhna hai ki india ki life expectancy over the years kaise change hui h
# x,y,aur data teeno paas karna zarrori h
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using relplot (Figure level) - line plot ko relplot use krke hi bana rhe h-- 'kind' pass karna padega
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
# mai bss india brazil aur germany ka life expectancy plot dekhna h
#sbse pehle hum filter kar rhe h data ko inn 3 countries k basis pe

temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
# figure level function k plot mai autamatically label plot k bahar aa jaaa rha h
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
# ye axes level plot h ---- isme label plot k andar hi plot ho rha h
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
#style parameter in relplot
# hume ye janana h ki konsi country konse continent ko belong karti h?
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country',style = 'continent') #style of line har continent k liye alag ho jayengi


#size parameter
#continent k basis pe line k size ko bhi change kar sakte h
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country',style = 'continent',size='continent') #har continent k according size of lines alag ho jayega
```

**FACET plot**

```python
# facet plot -> figure level function -> it will work with sns.relplot function kyunki ye figure level plot h
# it will not work with sns.scatterplot function and lineplot kyunki ye axes level plots h
#row or column or both paramenters ko pass kar do
#kind mai line aur scatter dono chalega
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter') ## saare dino k basis pe alag alag plots aa jayenge
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex') # sex k basis pe differentiation alag plots aa jayenge
```

```python
#facet grid plot
# suppose hume male and female ka variation alag alag (side by side graph) mai chahiye (alag alag column mai chahiye --> parameter used : 'col')

sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex') ## male aur female ka variation for total bill vs tip side by side aa gye
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex') ## male aur female ka variation for total bill vs tip row by row dekhna ho toh
```

```python
#kind= 'line' ya 'scatter' dono chalega
# ## rows wise hum dekhana chahte h ki sex wise variation kaisa h aur column wise hum dekhna chahte h ki smoker h ya nhi
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex', col='smoker') 
```

```python
# suppose abb hume day wise sex ka variation dekhna h total bill vs tip mai
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='day', col='sex') 
```

```python
#upar wala plot as a line plot
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', row='day', col='sex') 
```

```python
# col wrap - parameter
gap
```

```python
#suppose mujhe har year k liye alg plot chahiye - column wise
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year')
```

```python
# upar wala curve kuch zyada kaam ka nhi h kyunki ek hi column mai bohot saare graphs plot ho gye h
```

```python
#lekin upar ka plot readable bhi nhi h kyunki bohot saare plots bann gye h wo bhi ek hi line mai
#col_wrap ek line mai sirf n number of plot hi banata h (jo number hum provide karte h col_wrap parameter ko)
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

### 1. Histplot - Histogram

```python
# plotting univariate histogram
# khud hi bin size create kar lega

# axis level -> histplot

sns.histplot(data=tips, x='total_bill') #univariate analysis
```

```python
# figure level

sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter - hum apne hisaab se bins ka number de sakte h

sns.displot(data=tips, x='total_bill', kind='hist',bins=20)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# countplot (not actually a histogram) - basically count plot hi h, distplot nhi h
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter

# male or female ka comarison kar rhe h
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
```

```python
# element -> step -- thoda acha aesthetic ka plot (upar wala hi h, bss sundar h)

sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step')
```

```python
# Titanic wala dataset
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x = 'age', kind='hist')
```

```python
# male or female ka age distribution dekh sakte h

sns.displot(data=titanic, x='age', kind='hist',element='step',hue='sex')
```

```python
# facet plots ka use ho sakta h, using col and row kyunki ye ek figure level function h-> does not work on sns.histplot function kyunki wo ek axes level function h

sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

```python
# kdeplot - kernel density estimator
# Rather than using discrete bins, a KDE plot smoothens the observations with a Gaussian kernel, producing a continuous density estimate
# y-axis pe probability h --> matlab iss plot k according ye bola jaa sakta h ki total_bill=20 ki iss dataset mai hone ki probability 0.05 = 5% h

sns.kdeplot(data=tips,x='total_bill')
```

```python
# figure level
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue and fill parameter
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# faaltu h height aur aspect parameter

sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip') #axes level
sns.displot(data=tips, x='total_bill', y='tip',kind='hist') #figure level

## faaltu plot h - kabhi use mat karna
```

```python
# Bivariate Kdeplot

# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')

#jitni close lines hongi utna hi zyada probability hogi unn values k observations ki uss region of values k close
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

#har country ka har saal life expectancy kitna tha, agar hume wo check karna h...?

# Plot rectangular data as a color-encoded matrix - pehle hume data ko pivot (pivot table) mai convert karna padega
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function (heatmap ka koi figure level function nhi hota)
plt.figure(figsize=(15,15))
sns.heatmap(temp_df) #axes level function ka plot h, aur iska koi aur alternate nhi h (koi figure level function se heatmap ko plot nhi kiya jaa sakta)
```

```python
# annot parameter - annotation (values bhi provide kar sakte h) provide karne k liye
# cmap parameter - color coding change kar sakte h
#linewidth - ye humare plot k har row mai thoda gap create kar dega

temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

# agar hume ye pata karna h ki humare given 4 columns mai konse columns zyada similar h (basically similarity pata karne k liye kaam aata h)

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])

# heatmap + dendogram
# isne automatically hi columns ka order change kar diya
# iss plot k according sepal width aur petal length sabse zyada similar h, aur petal width and sepal length sabse zyada dissimilar h
# and similarly usne row k sath bhi yahi kiya h
```

```python
#how to change the size of a figure level plot = using 'height' and 'aspect' parameter
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```
